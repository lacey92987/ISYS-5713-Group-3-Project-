Compare two heroes

```python


def compare_heroes(hero_id1, hero_id2):
    url = 'http://127.0.0.1:5000/compare_power'
    params = {'hero_id1': hero_id1, 'hero_id2': hero_id2}
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        result = response.json()
        print(f"Comparison Result:")
        print(f"{result['hero_name1']} (ID: {result['hero_id1']}) - Total Power: {result['total_power1']}")
        print(f"vs.")
        print(f"{result['hero_name2']} (ID: {result['hero_id2']}) - Total Power: {result['total_power2']}")
        print(f"Winner: {result['winner']}")
    else:
        print(f"Error: {response.status_code} - {response.text}")

if __name__ == "__main__":
    hero_id1 = input("Enter ID of the first hero: ")
    hero_id2 = input("Enter ID of the second hero: ")

    compare_heroes(hero_id1, hero_id2)
```

Add a Hero

```python


def add_new_hero():
    url = 'http://127.0.0.1:5000/heroes'

    # Prompt user for hero details
    hero_name = input("Enter Hero Name: ")
    gender = input("Enter Gender: ")
    eye_color = input("Enter Eye Color: ")
    species = input("Enter Species: ")
    hair_color = input("Enter Hair Color: ")
    height = float(input("Enter Height: "))
    weight = float(input("Enter Weight: "))
    publisher = input("Enter Publisher: ")
    skin_color = input("Enter Skin Color: ")
    alignment = input("Enter Alignment: ")

    # Prepare data for POST request
    data = {
        "hero_name": hero_name,
        "gender": gender,
        "eye_color": eye_color,
        "species": species,
        "hair_color": hair_color,
        "height": height,
        "weight": weight,
        "publisher": publisher,
        "skin_color": skin_color,
        "alignment": alignment
    }

    # Make POST request
    response = requests.post(url, json=data)

    if response.status_code == 201:
        hero_id = response.json().get('hero_id')
        print(f"Hero added successfully with ID: {hero_id}")
    else:
        print(f"Error: {response.status_code} - {response.text}")

if __name__ == "__main__":
    add_new_hero()
```

Delete a hero

```python


def delete_hero(hero_id):
    url = f'http://127.0.0.1:5000/heroes/{hero_id}'

    # Make DELETE request
    response = requests.delete(url)

    if response.status_code == 200:
        print(f"Hero with ID {hero_id} deleted successfully.")
    elif response.status_code == 404:
        print(f"Error: Hero with ID {hero_id} not found.")
    else:
        print(f"Error: {response.status_code} - {response.text}")

if __name__ == "__main__":
    hero_id_to_delete = input("Enter the ID of the hero you want to delete: ")

    try:
        hero_id_to_delete = int(hero_id_to_delete)
        delete_hero(hero_id_to_delete)
    except ValueError:
        print("Invalid hero ID. Please enter a valid numeric ID.")
```

Update a hero

```python


def update_hero(hero_id, updated_data):
    url = f'http://127.0.0.1:5000/heroes/{hero_id}'

    # Make PUT request
    response = requests.put(url, json=updated_data)

    if response.status_code == 200:
        updated_hero = response.json().get('hero')
        print(f"Hero with ID {hero_id} updated successfully.")
        print("Updated Hero Details:")
        print(updated_hero)
    elif response.status_code == 404:
        print(f"Error: Hero with ID {hero_id} not found.")
    else:
        print(f"Error: {response.status_code} - {response.text}")

if __name__ == "__main__":
    hero_id_to_update = input("Enter the ID of the hero you want to update: ")

    try:
        hero_id_to_update = int(hero_id_to_update)
        
        # Prompt user for updated hero details
        updated_hero_data = {
            "hero_name": input("Enter Updated Hero Name: "),
            "gender": input("Enter Updated Gender: "),
            "eye_color": input("Enter Updated Eye Color: "),
            "species": input("Enter Updated Species: "),
            "hair_color": input("Enter Updated Hair Color: "),
            "height": float(input("Enter Updated Height: ")),
            "weight": float(input("Enter Updated Weight: ")),
            "publisher": input("Enter Updated Publisher: "),
            "skin_color": input("Enter Updated Skin Color: "),
            "alignment": input("Enter Updated Alignment: ")
        }

        update_hero(hero_id_to_update, updated_hero_data)
    except ValueError:
        print("Invalid hero ID. Please enter a valid numeric ID.")
```
