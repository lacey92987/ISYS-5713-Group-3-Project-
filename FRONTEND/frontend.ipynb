The `url` variable can be edited if required.

```python
# Set base URL for API calls
url = 'http://localhost:5000/'

# We'll use this to append to the base URL for specific API calls
url_part = ''

# We'll reserve the payload for POST/PUT methods generally
payload = {}

# We don't have any specific headers, but this can be used for authentication if required
headers = {}
```

## Basic Calls

```python
# Get all heroes

response = requests.request("GET", url + 'heroes', headers=headers, data=payload)
heroes = response.json()

print(f"Respose Status: {response.status_code}")
heroes
```

```python
# Get powers for a specific hero

hero_id = input("Enter hero ID: ")
url_part = f'heroes/{hero_id}/powers'
response = requests.request("GET", url + url_part, headers=headers, data=payload)
powers = response.json()

print(f"Respose Status: {response.status_code}")
powers
```
